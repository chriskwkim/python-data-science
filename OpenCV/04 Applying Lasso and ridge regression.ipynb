```python
import numpy as np
from sklearn import linear_model
from sklearn import model_selection as model
from sklearn import datasets
from sklearn import metrics
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
linreg = linear_model.LinearRegression()
```

```python
lassoreg = linear_model.Lasso()
```

```python
ridgereg = linear_model.Ridge() # RidgeRegression()
```

```python
boston = datasets.load_boston()
```

```python
X_train, X_test, y_train, y_test = model.train_test_split(boston.data,
                                                         boston.target,
                                                         test_size=0.1,
                                                         random_state=42)
```

```python
linreg.fit(X_train, y_train)
lassoreg.fit(X_train, y_train)
ridgereg.fit(X_train, y_train)
```

```python
mse_lin = metrics.mean_squared_error(y_train, linreg.predict(X_train))
mse_lasso = metrics.mean_squared_error(y_train, lassoreg.predict(X_train))
mse_ridge = metrics.mean_squared_error(y_train, ridgereg.predict(X_train))

print("MSE Linear Reg: {}, Lasso: {}, Ridge: {}".format(mse_lin, mse_lasso, mse_ridge))
```

```python
print("Linear Reg Score: {}".format(linreg.score(X_train, y_train)))
print("Lasso Reg Score : {}".format(lassoreg.score(X_train, y_train)))
print("Ridge Reg Score : {}".format(ridgereg.score(X_train, y_train)))
```

```python
y_pred_lin = linreg.predict(X_test)
y_pred_lasso = lassoreg.predict(X_test)
y_pred_ridge = ridgereg.predict(X_test)

print("MSE test Linear Reg: {}".format(metrics.mean_squared_error(y_test, y_pred_lin)))
print("MSE test Lasso Reg : {}".format(metrics.mean_squared_error(y_test, y_pred_lasso)))
print("MSE test Ridge Reg : {}".format(metrics.mean_squared_error(y_test, y_pred_ridge)))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y_test, linewidth=3, label='ground truth')
plt.plot(y_pred_lin, linewidth=3, label='linear')
plt.plot(y_pred_lasso, linewidth=3, label='lasso')
plt.plot(y_pred_ridge, linewidth=3, label='ridge')
plt.legend(loc='best')
plt.xlabel('test data points')
plt.ylabel('target values')
```

```python
plt.plot(y_test, y_pred_lin, 'o')
plt.plot([-10, 60], [-10, 60], 'k--')
plt.axis([-10, 60, -10, 60])
plt.xlabel('ground truth')
plt.ylabel('linear reg predicted')
```
